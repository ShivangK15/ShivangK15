1. Merge the datasets Customers, Product Hierarchy and Transactions as Customer_Final. Ensure to
keep all customers who have done transactions with us and select the join type accordingly.

```python
Customer_Transactions=pd.merge(Customer,Transactions,left_on="customer_Id", right_on="cust_id",how='inner')
```

```python
Customer_Transactions
```

```python
Customer_Final=pd.merge(Customer_Transactions,prod_cat_info,left_on=['prod_cat_code','prod_subcat_code'],right_on=['prod_cat_code','prod_sub_cat_code'],how='left')
```

3. Generate histograms for all continuous variables and frequency bars for categorical variables.

```python
cont_vars = ['Qty','Rate','Tax','total_amt']
cat_vars = ['Gender','city_code','Store_type','prod_cat','prod_subcat']
for i in cont_vars:
    plt.figure(figsize=(5,5))
    sns.histplot(Customer_Final[i], bins=20)
    plt.title('Histogram for '+i)
    plt.xlabel(i)
    plt.ylabel('Frequency')
    plt.show()
```

```python
for j in cat_vars:
    plt.figure(figsize=(6, 6))
    sns.countplot(x=j, data=Customer_Final)
    plt.title('Count Plot for ' + j)
    plt.xlabel(j)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

9. What was the total amount earned from "Male" customers under the "Electronics" category?

```python
sum(Customer_Final.loc[(Customer_Final['Gender']=='M')&(Customer_Final['prod_cat']=='Electronics'),'total_amt'])
```

10. How many customers have more than 10 unique transactions, after removing all transactions which have any negative amounts?

```python
unique_postrans=Customer_Final.loc[Customer_Final['total_amt']>=0,:].groupby('customer_Id').transaction_id.nunique().reset_index()
```

```python
n_customers=len(unique_postrans[unique_postrans['transaction_id']>10])
print("The number of customers that have more than 10 unique transactions is:",n_customers)
```

b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?

```python
filtered_cus=Customer_Final.loc[(Customer_Final['Age']>=25)&(Customer_Final['Age']>=35),:]
```

```python
amount=sum(filtered_cus.loc[(filtered_cus['tran_date']>='2014-01-01')&(filtered_cus['tran_date']<='2014-03-01'),'total_amt'])
```

```python
print("The total amount spent by the customers of age between 25 and 35 between 1st Jan, 2014 to 1st Mar, 2014 is:",amount)
```
